```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_classification
```

```python
X,y = make_classification(n_samples=1000, n_features=20,n_informative=3,n_classes = 3,random_state=42)
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.30,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(multi_class='ovr')
lr.fit(X_train,y_train)
```

```python
y_pred = lr.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
accuracy_score(y_test,y_pred)
confusion_matrix(y_test,y_pred)
# print(classification_report(y_test,y_pred))
# print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
```

# Task
Tune the hyperparameters of the Logistic Regression model using GridSearchCV to find the best 'C' and 'solver' values. Then, train a new model with these optimal parameters, predict on the test set, and evaluate its accuracy, confusion matrix, and classification report, comparing the new accuracy with the previous one.

## Define Hyperparameter Grid

### Subtask:
Define a dictionary with hyperparameters for Logistic Regression, such as 'C' (regularization strength) and 'solver', to explore during tuning.


**Reasoning**:
I will define a dictionary `param_grid` containing the specified hyperparameters ('C' and 'solver') and their respective lists of values for Logistic Regression tuning.



```python
param_grid = {
    'C': [0.1, 1, 10, 100],
    'solver': ['lbfgs', 'liblinear', 'newton-cg', 'sag', 'saga']
}
print(param_grid)
```

**Reasoning**:
Now that the hyperparameter grid is defined, the next step is to import the `GridSearchCV` class from `sklearn.model_selection` to perform an exhaustive search over the specified parameter values for the Logistic Regression model.



```python
from sklearn.model_selection import GridSearchCV
print("GridSearchCV imported successfully.")
```

**Reasoning**:
To perform hyperparameter tuning, I will initialize `GridSearchCV` with a `LogisticRegression` estimator, the `param_grid` containing the hyperparameters to search, and set `cv=5` for cross-validation to find the optimal combination of parameters.



```python
logreg = LogisticRegression(multi_class='ovr', solver='liblinear', max_iter=1000) # Setting solver to liblinear and max_iter for broad compatibility across C values
grid_search = GridSearchCV(logreg, param_grid, cv=5, scoring='accuracy', verbose=1, n_jobs=-1)
grid_search.fit(X_train, y_train)
print("GridSearchCV initialized and fitted.")
```

**Reasoning**:
Now that GridSearchCV has completed fitting, the next logical step is to retrieve and display the best hyperparameters and their corresponding accuracy score found during the search.



```python
print(f"Best parameters: {grid_search.best_params_}")
print(f"Best cross-validation accuracy: {grid_search.best_score_:.4f}")
```

**Reasoning**:
With the best hyperparameters identified, the next step is to train a new Logistic Regression model using these optimal parameters. This model will then be used to make predictions on the test set.



```python
best_lr = grid_search.best_estimator_
print(f"Best Logistic Regression model: {best_lr}")
```

**Reasoning**:
Now that we have the best model from GridSearchCV, the next step is to use this `best_lr` model to make predictions on the unseen test data (`X_test`).



```python
y_pred_tuned = best_lr.predict(X_test)
print(y_pred_tuned)
```

**Reasoning**:
Now that predictions from the tuned model (`y_pred_tuned`) are available, the next logical step is to evaluate its performance by calculating the accuracy score against the true test labels (`y_test`).



```python
tuned_accuracy = accuracy_score(y_test, y_pred_tuned)
print(f"Tuned Model Accuracy: {tuned_accuracy:.4f}")
```

**Reasoning**:
With the tuned model's predictions, the next step is to evaluate its performance more comprehensively by generating and printing the confusion matrix, which provides a detailed breakdown of correct and incorrect classifications for each class.



```python
tuned_confusion_matrix = confusion_matrix(y_test, y_pred_tuned)
print("Tuned Model Confusion Matrix:")
print(tuned_confusion_matrix)
```

**Reasoning**:
To get a more detailed evaluation of the tuned model's performance, I will generate and print the classification report, which includes precision, recall, and F1-score for each class.



```python
tuned_classification_report = classification_report(y_test, y_pred_tuned)
print("Tuned Model Classification Report:")
print(tuned_classification_report)
```

**Reasoning**:
All evaluation metrics for the tuned model have been generated. The final step of the task is to compare the accuracy of the tuned model with the initial model and then mark the subtask as complete.



```python
print(f"Initial Model Accuracy: {accuracy_score(y_test, y_pred):.4f}")
print(f"Tuned Model Accuracy: {tuned_accuracy:.4f}")

if tuned_accuracy > accuracy_score(y_test, y_pred):
    print("The tuned model has a higher accuracy.")
elif tuned_accuracy < accuracy_score(y_test, y_pred):
    print("The tuned model has a lower accuracy.")
else:
    print("The tuned model has the same accuracy as the initial model.")
```
